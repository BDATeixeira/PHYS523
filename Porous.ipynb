```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.datasets import make_circles
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection
from matplotlib import cm
from matplotlib import animation
import math as mt
```

```python
# put circles on a box
# no overlapping circles
# aim to 55% fill
# circle radius between 0.02-0.1
```

```python
def myradius(t):
    t is int
    l=np.random.uniform(0.02,0.1,t) #return random float in interval
    return l                                 
```

```python
coordenada_x=[np.random.rand(1)] #lista coordenada x do circulo
coordenada_y=[np.random.rand(1)] #lista coordenada y do circulo
coordenada_r=[myradius(1)] #lista dimensao raio do circulo
#circulos=[] #lista de circulos
Area=0 #contador de area

while (Area<0.65): #definir percentagem a ser preenchida pelo ciclo while
    x=np.random.rand(1) #criar o primeiro circulo
    y=np.random.rand(1)
    r=myradius(1)
    t=0   # contador distancia entre circulos
  
    for a,b,c in zip(coordenada_x,coordenada_y,coordenada_r): #funcao zip a emparelhar os elementos de cada lista
        d=np.sqrt(((x-a)**2)+((y-b)**2)) #verificar distancia ente circulos
        if d<(r+c): #loop if de distancia, para ter a certeza que nao se tocam
            t+=1
            
    if t==0:   #quando nao se tocam escrever as coordenadas nas listas (append)
        coordenada_x.append(x)
        coordenada_y.append(y)
        coordenada_r.append(r)
        Area+=(np.pi)*(r**2) #quando nao se tocam adiciona ao contador de area

fig,ax=plt.subplots() #subplot
ax.set_aspect(1) # definir que eixo dos xx e yy esta igual
ax.set_facecolor('dimgrey') #cor de background
ax.set(xlim=(0,1), ylim=(0,1)) #definir limites dos eixos
for i in range(len(coordenada_x)): #loop for para cada ponto dentro da lista (todas as listas tem a mm dimensao!)
    circle = plt.Circle((coordenada_x[i], coordenada_y[i]), coordenada_r[i],color='bisque') #fazer plot de um circulo para cada emparelhamento de pontos
    ax.add_artist(circle)  #plot circulo   
plt.show()
```
